##### No m'interesen els valors NaN per tant, faré una mitjana del barri i aquesta serà assignada al valor NaN. Però, abans d'això farè un canvi més al df ja que em poden sortir mitjanes erronies. 

##### El següent pas és dividir el df en 2. Si s'observa amb atenció el df complert, es pot veure com en la columna preu hi han dos tipus de dades, el preu del lloguer mensula mitjà i el preu mitjà per m2 mensual. Amb la funció .unique() es veu millor:

```python
df["Lloguer_mitja"].unique() 
```

```python
#Creo dos df nous a partir del df.
llog_m = df.loc[df["Lloguer_mitja"] == "Lloguer mitjà mensual (Euros/mes)"]
llog_m_sup = df.loc[df["Lloguer_mitja"] == "Lloguer mitjà per superfície (Euros/m2 mes)"]
```

##### Ara sí, canvio els valors NaN per la mitjana del seu barri.

```python
#Calculo la mitjana per cada barri de Barcelona
llog_m_barri = llog_m.groupby('Nom_Barri')['Preu'].mean()
llog_m_barri
```

##### Com que el barri Can Peguera no pot fer mitjana ja que només conté valos nulls, agafem la mitjana del districte.

```python
llog_m_districte = llog_m.groupby('Nom_Districte')['Preu'].mean()
llog_m_districte
```

```python
llog_m['Preu'] = llog_m['Preu'].fillna(llog_m["Nom_Districte"].map(llog_m_districte))
```

```python
llog_m.isnull().sum() # Ara ja no hi han valors nulls al df llog_m
```
